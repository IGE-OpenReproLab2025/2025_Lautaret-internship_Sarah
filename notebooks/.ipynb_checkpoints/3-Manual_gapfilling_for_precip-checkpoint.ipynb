**Date** : 30/04/2025 \
**Biblio used** : Vionnet, V., Brun, E., Morin, S., Boone, A., Faroux, S., Le Moigne, P., Martin, E., and Willemet, J.-M.: The detailed snowpack scheme Crocus and its implementation in SURFEX v7.2, Geosci. Model Dev., 5, 773â€“791, https://doi.org/10.5194/gmd-5-773-2012, 2012. \
**Practical steps** :
1. Read automatically filled lautaret's data + read MeteoFrance sattion's data
2. Locate dates to fill
3. Fill dates with zero precip (checking with lautaret's webcam)
4. For isolated snow event with missing data : use of crocus formula
5. For multiple events with snow and rain : use a mean of nearby meteofrance stations
6. Correction of a missing event : 13 march 2022 (rain gauge overflow)
7. Write filled data in a file

### Fill gaps where there is snow precipitation
**Method** : use crocus formula to calculate the density of freshly fallen snow and back up to cumulative precipitation of an event

#### 15 Dec 2020

```python
## Episod from 15/12 3:00 to 15/12 14:30
density_crocus = fct.densite_crocus(df.loc['2020-12-15 3:00:00' :'2020-12-15 14:30:00' , 'AirTC_Avg'].mean(), df.loc['2020-12-15 3:00:00' :'2020-12-15 14:30:00', 'WindSpeed_Avg'].mean())
snow_acc = df.loc['2020-12-15 14:30:00', 'Snow_Depth'] - df.loc['2020-12-15 3:00:00', 'Snow_Depth']
precip_corrected = density_crocus * snow_acc
remaining_precip = precip_corrected - df.loc['2020-12-15 3:00:00' :'2020-12-15 14:30:00', 'Quantity_raw'].sum()

print("The cumulated precipitation estimated for all the event thanks to the crocus density's formula is "+str(round(precip_corrected,2))+" mm, for a snow density of "+str(round(density_crocus, 2))+" kg/m3")
print("The remaining cumulated precip to fill is "+str(round(remaining_precip,2))+" mm on "+str(df.loc['2020-12-15 04:00:00', 'len_gap_Quantity_raw']/2)+' h')
```

```python
serie_to_fill = df.loc['2020-12-15', 'Quantity_raw']

# Annotation code for gapfilling : 8 for using density from crocus formula to get precipitation amout 
df_gapfilling_annotation.loc['2020-12-15', 'Quantity_raw'].loc[df['Quantity_raw'].isna()] = 8

serie_to_fill.loc[df['Quantity_raw'].isna()] = remaining_precip/len(serie_to_fill.loc[df['Quantity_raw'].isna()])
```

```python
days = ['2020-10-02', '2020-10-03', '2020-10-04', '2020-10-05', '2020-10-06', '2020-10-07', '2020-10-10', '2020-10-11']

for i in days :
    mean = (df_MF.loc[i, 'RR_besse']+df_MF.loc[i, 'RR_valloire'] + df_MF.loc[i, 'RR_monetier'])/3
    missing_precip = mean - df.loc[i, 'Quantity_raw'].sum()
    serie_to_fill = df.loc[i, 'Quantity_raw']

    # Annotation code for gapfilling : 3 for using meteofrance stations nearby 
    df_gapfilling_annotation.loc[i, 'Quantity_raw'].loc[df['Quantity_raw'].isna()] = 3
    
    serie_to_fill.loc[df['Quantity_raw'].isna()] = missing_precip/len(serie_to_fill.loc[df['Quantity_raw'].isna()])
```
