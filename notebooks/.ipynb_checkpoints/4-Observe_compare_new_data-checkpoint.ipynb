## Comparison precipitation using crocus density and apparent density

### Snow accumulation calculation

```python
df_raw['Snow_acc'] = df_raw['Snow_Depth'].diff()
df_raw['Snow_acc'] = df_raw.loc[df_raw['Snow_acc']>0, 'Snow_acc']
```

### Resample data by day

```python
# Creation of a new df with daily data
df_raw_1d = pd.DataFrame()

# Density calculation with Flux'Alp data
df_raw_1d['Snow_acc'] = df_raw['Snow_acc'].resample('d').sum()
df_raw_1d['Quantity_raw'] = df_raw['Quantity_raw'].resample('d').sum()
df_raw_1d['density'] = df_raw_1d['Quantity_raw']/df_raw_1d['Snow_acc']
df_raw_1d['density'] = df_raw_1d.loc[(df_raw['Quantity_raw']>0) & (df_raw['AirTC_Avg']<1), 'density']  # Only keeping days with precipitations and Tmean < 1Â°C

# Density calculation with Crocus formula
df_raw_1d['AirTC_Avg'] = df_raw['AirTC_Avg'].resample('d').mean()
df_raw_1d['WindSpeed_Avg'] = df_raw['WindSpeed_Avg'].resample('d').mean()
df_raw_1d['density_crocus'] = fct.densite_crocus(df_raw_1d['AirTC_Avg'], df_raw_1d['WindSpeed_Avg'])
df_raw_1d['density_crocus'] = df_raw_1d.loc[(df_raw['Quantity_raw']>0) & (df_raw['AirTC_Avg']<1), 'density_crocus']  # Only keeping days with precipitations

# Precip calculation with Crocus formula
df_raw_1d['precip_crocus'] = df_raw_1d['density_crocus']*df_raw_1d['Snow_acc']
df_raw_1d['prepic_fluxalp'] = df_raw_1d.loc[(df_raw['Quantity_raw']>0) & (df_raw['AirTC_Avg']<1), 'Quantity_raw'] # Only keeping days with precipitations
```

### Plot density

```python
# Plot Density
fig, ax = plt.subplots(2,2,figsize=(15,8))

df_raw_1d.loc['2020-11' : '2021-04','density_crocus'].plot(ax = ax[0,0], style='.', label='Crocus density', color = '#9a57d2')
df_raw_1d.loc['2020-11' : '2021-04','density'].plot(ax = ax[0,0], style='.', label = "Flux'Alp density", color = '#3d57bd')

df_raw_1d.loc['2021-11' : '2022-04','density_crocus'].plot(ax = ax[0,1], style='.', label='Crocus density', color = '#9a57d2')
df_raw_1d.loc['2021-11' : '2022-04','density'].plot(ax = ax[0,1], style='.', label = "Flux'Alp density", color = '#3d57bd')

df_raw_1d.loc['2022-11' : '2023-04','density_crocus'].plot(ax = ax[1,0], style='.', label='Crocus density', color = '#9a57d2')
df_raw_1d.loc['2022-11' : '2023-04','density'].plot(ax = ax[1,0], style='.', label = "Flux'Alp density", color = '#3d57bd')

df_raw_1d.loc['2023-11' : '2024-04','density_crocus'].plot(ax = ax[1,1], style='.', label='Crocus density', color = '#9a57d2')
df_raw_1d.loc['2023-11' : '2024-04','density'].plot(ax = ax[1,1], style='.', label = "Flux'Alp density", color = '#3d57bd')

for i in range(2) :
    for j in range(2):
        ax[i,j].legend(loc ="upper right")
        ax[i,j].set_ylim(0,200)
        ax[i,j].set_xlabel("")
        ax[i,j].set_ylabel("Density (kg/m3)")
        ax[i,j].grid(color='grey', linestyle=':', linewidth=1)
```

### Plot precipitation

```python
fig, ax = plt.subplots(2,2,figsize=(15,8))

df_raw_1d.loc['2020-11' : '2021-04','precip_crocus'].plot(ax = ax[0,0], style='.', label='Crocus precip', color = '#9a57d2')
df_raw_1d.loc['2020-11' : '2021-04','prepic_fluxalp'].plot(ax = ax[0,0], style='.', label = "Flux'Alp precip", color = '#3d57bd')

df_raw_1d.loc['2021-11' : '2022-04','precip_crocus'].plot(ax = ax[0,1], style='.', label='Crocus precip', color = '#9a57d2')
df_raw_1d.loc['2021-11' : '2022-04','prepic_fluxalp'].plot(ax = ax[0,1], style='.', label = "Flux'Alp precip", color = '#3d57bd')

df_raw_1d.loc['2022-11' : '2023-04','precip_crocus'].plot(ax = ax[1,0], style='.', label='Crocus precip', color = '#9a57d2')
df_raw_1d.loc['2022-11' : '2023-04','prepic_fluxalp'].plot(ax = ax[1,0], style='.', label = "Flux'Alp precip", color = '#3d57bd')

df_raw_1d.loc['2023-11' : '2024-04','precip_crocus'].plot(ax = ax[1,1], style='.', label='Crocus precip', color = '#9a57d2')
df_raw_1d.loc['2023-11' : '2024-04','prepic_fluxalp'].plot(ax = ax[1,1], style='.', label = "Flux'Alp precip", color = '#3d57bd')

for i in range(2) :
    for j in range(2):
        ax[i,j].legend(loc ="upper right")
        ax[i,j].set_ylim(0,90)
        ax[i,j].set_xlabel("")
        ax[i,j].set_ylabel("Precipitation (mm/days)")
        ax[i,j].grid(color='grey', linestyle=':', linewidth=1)
```

### Comparison both precipitation's data

```python
plt.plot(df_raw_1d['precip_crocus'], df_raw_1d['prepic_fluxalp'], '.')
plt.xlabel('Crocus precipitation (mm/d)')
plt.ylabel("Flux'alp precipitation (mm/d)")
plt.axline([0,0], slope = 1, color = 'black', label = 'x=y')
plt.grid(color='grey', linestyle=':', linewidth=1)
plt.legend()
```

### Monthly stats calculation

```python
stats = pd.DataFrame()

stats['cumul_crocus_precip'] = df_raw_1d['precip_crocus'].resample('MS').sum()
stats['cumul_fluxalp_precip'] = df_raw_1d['prepic_fluxalp'].resample('MS').sum()
stats['mean_crocus_precip'] = df_raw_1d['precip_crocus'].resample('MS').mean()
stats['mean_fluxalp_precip'] = df_raw_1d['prepic_fluxalp'].resample('MS').mean()
```

### Plot monthly cumulative and mean precipitation

```python
fig, ax = plt.subplots(1,2,figsize=(15,5), sharex=True)

stats['cumul_crocus_precip'].plot(ax = ax[0], label = 'Crocus precipitation', color = '#9a57d2')
stats['cumul_fluxalp_precip'].plot(ax = ax[0], label = "Flux'Alp precipitation", color = '#3d57bd')
ax[0].set_title('Monthly cumulative precipitation')



stats['mean_crocus_precip'].plot(ax = ax[1], label = 'Crocus precipitation', color = '#9a57d2')
stats['mean_fluxalp_precip'].plot(ax = ax[1], label = "Flux'Alp precipitation", color = '#3d57bd')
ax[1].set_title('Monthly mean precipitation')


for ca in ax :
    ca.legend()
    ca.set_ylabel('Precipitation (mm)')
    ca.set_xlabel('')
    ca.grid(color='grey', linestyle=':', linewidth=1)
```

### Biais calculation

```python
mean_biais = (stats['mean_crocus_precip'] - stats['mean_fluxalp_precip']).mean()
```

```python
print('The monthly mean of the daily cumulative precipitation is overestimated by an average of '+str(round(mean_biais, 2))+' mm/d with the crocus formula.')
```
