**Date** : 30/04/2025 \
**Data used** : Lautaret's data : *https://gricad-gitlab.univ-grenoble-alpes.fr/lautaret/fluxalp/-/tree/master/data_clean_hydro_year?ref_type=heads* \
**Practical steps** :
1. Read data
2. Unit conversion
3. Conversion relative humidity to specific humidity
4. Adding a column of zeros for y-wind-speed
5. Write data for a Parflow forcing file
6. Visualize forcing data

## Import

```python
import sys
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sys.path.append("../scripts/")
import file_functions as fct
```

## Read filled lautaret's data

```python
df_filled = pd.read_csv('resulting_data/Fluxalp_filled_2018_2024.dat')
df_filled['datetime'] = pd.to_datetime(df_filled['datetime'])
df_filled.set_index('datetime', inplace=True)

var_for_parflow = ['short_up_Avg', 'long_up_cor_Avg', 'Quantity_raw', 'AirTC_Avg', 'WindSpeed_Avg', 'Patm_Avg',  'HRair_Avg']
df = df_filled.copy().loc[:,var_for_parflow]
df.head()
```

## Unit conversion

```python
df['Quantity_raw'] = df['Quantity_raw'].div(1800)      ## mm/30min to mm/s
df['AirTC_Avg'] = df['AirTC_Avg'].add(273.15)          ## Â°C to K
df['Patm_Avg'] = df['Patm_Avg'].mul(100)               ## hPa to Pa
```

## Relative humidity to specific humidity

```python
df['Specific_H'] = fct.convert_relativH_to_speH(df['HRair_Avg'], df['AirTC_Avg'], df['Patm_Avg'])
```

## Adding a column for y-windspeed

```python
df['wind_speed_y'] = 0.0
```

## Checking for missing value

```python
for i in df.columns :
    print(str(df[i].isna().sum())+' missing value for '+i)
```

## Writing data in a file

```python
df = df[['short_up_Avg', 'long_up_cor_Avg', 'Quantity_raw', 'AirTC_Avg', 'WindSpeed_Avg', 'wind_speed_y', 'Patm_Avg', 'Specific_H']]
df.head()
```

```python
output_file = df.copy()

## Saved at the same place
output_file.to_csv('resulting_data/Parflow_forcing_2018_2024.dat', sep = '\t', index=False, header=False, float_format = "%5f")

## Saved in my private storage
path = '/home/jovyan/private-storage/input_parflow/'
output_file.to_csv(path+'/Parflow_forcing_2018_2024.dat', sep = '\t', index=False, header=False, float_format = "%5f")
```

## Create input file for each hydro year

```python
start_year = ['2018', '2019', '2020', '2021', '2022', '2023']
end_year =['2019', '2020', '2021', '2022', '2023', '2024']

for i,x in enumerate(start_year):
    df_year = df.loc[x+'-10-17' : end_year[i]+'-10-16']

    ## Saved at the same place
    df_year.to_csv('resulting_data/Parflow_forcing_'+x+'_'+end_year[i]+'.dat', sep = '\t', index=False, header=False, float_format = "%5f")

    ## Saved in my private storage
    df_year.to_csv(path+'/Parflow_forcing_'+x+'_'+end_year[i]+'.dat', sep = '\t', index=False, header=False, float_format = "%5f")
```
